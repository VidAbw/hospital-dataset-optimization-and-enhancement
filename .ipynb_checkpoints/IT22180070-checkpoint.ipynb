```python
import pandas as pd
import numpy as np

# දත්ත කියවීම
df = pd.read_csv('diabetes.csv')

# '?' ලකුණු NaN බවට පත් කිරීම සහ අනවශ්‍ය columns ඉවත් කිරීම
df.replace('?', np.nan, inplace=True)
df.drop(columns=['weight', 'payer_code', 'medical_specialty', 'id', 'encounter_id', 'patient_nbr'], inplace=True)

# හිස්තැන් (Missing values) පිරවීම
for col in df.columns:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].mode()[0], inplace=True)

print("දත්ත සාර්ථකව Load කර පිරිසිදු කර ඇත. දැන් ප්‍රස්තාර ඇඳිය හැක!")
```
